```python
xaxis = np.arange(30,70,1)
yaxis = norm.pdf(xaxis,50,5)
plt.plot(xaxis,yaxis)
plt.axvline(50,color='k',linestyle='dashed')
plt.show()
```

```python
#Now forget about mean and std dev
#Sampling of data with different sizes
```

```python
np.random.seed(1)
sizes = []
for x  in range(10,20000,200):
    sizes.append(x)
means = [np.mean(5*np.random.randn(size)+50) for size in sizes]
```

```python
plt.scatter(sizes,np.array(means)-50)
plt.hlines(y = 0,xmin=0,xmax=20000,colors='k')
```

### Density Function

```python
#plot the gaussian pdf
from numpy import arange
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
sample_space = arange(-5,5,0.001)
mean = 0.0
stdev = 1.0
pdf = norm.pdf(sample_space,mean, stdev)
plt.plot(sample_space, pdf)
plt.axvline(mean, color='k', linestyle='dashed', linewidth=1)
plt.show()
```

### CDF

```python
cdf = norm.cdf(sample_space)
plt.plot(sample_space,cdf)
plt.show()
```
