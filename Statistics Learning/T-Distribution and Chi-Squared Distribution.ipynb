```python
from numpy import arange
from matplotlib import pyplot as plt
from scipy.stats import t
```

```python
sample_space = arange(-5,5,0.001)
dof = len(sample_space)-1
# In practical dof for t-dsitribution is number of observation -1
# As we are computing observation from other statistical distibution 
# in t- distribution, dof will be subtracted by 1
```

```python
pdf = t.pdf(sample_space,dof)
plt.plot(sample_space,pdf)
plt.show()
```

A key difference is the fatter tails in the
distribution (hard to see by eye), highlighting the increased likelihood of observations in the
tails compared to that of the Gaussian

```python
#cdf
cdf = t.cdf(sample_space,dof)
plt.plot(sample_space,cdf)
plt.show()
```

Running the example, we see the familiar S-shaped curve as we see with the Gaussian
distribution, although with slightly softer transitions from zero-probability to one-probability
for the fatter tails.

## Chi-Sqaured Distribution

```python
from scipy.stats import chi2
```

```python
sample_space = arange(0,50,0.01)
dof = 20
pdf = chi2.pdf(sample_space,dof)
```

```python
plt.plot(sample_space,pdf)
plt.show()
```

the distribution has a bell-like shape, the distribution is
not symmetric.

```python
cdf = chi2.cdf(sample_space,dof)
plt.plot(sample_space,cdf)
plt.show()
```

The distribution helps to see the likelihood for the Chi-Squared value around 20
with the fat tail to the right of the distribution that would continue on long after the end of the
plot.
