```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Line Plot

```python
x = [x*0.1 for x in range(100)]
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

## 2. Bar Plot

```python
import random
```

```python
random.seed(1)
x = ['Red', 'Green', 'Blue']
y = [random.randint(0,100), random.randint(0,100),random.randint(0,100)]
plt.bar(x,y)
plt.show()
```

## 3. Histogram Plot

```python
x = np.random.randn(1000)
plt.hist(x)
plt.axvline(np.mean(x),linestyle='dashed',color='k')
plt.show()
```

```python
plt.hist(x, bins=100)
plt.show()
```

## 4. Box and Whisker Plot

```python
random.seed(1)
x = [np.random.randn(1000),5*np.random.randn(1000)+50,
    10*np.random.randn(1000)]
plt.boxplot(x)
plt.show()
```

## 5. Scatter Plot

```python
random.seed(1)
x = 20*np.random.randn(1000) + 100
y = x + (10 * np.random.randn(1000) + 50)

plt.scatter(x,y)
plt.show()
```
